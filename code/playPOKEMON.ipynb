# Q-learning

```python
rlalg = QLearningAlgorithm
maxIter = 500000
# opponent does random actions
rlLevelOne = QLearning(rlalg, randomOppPolicy, identityFeatureExtractor, maxIter = maxIter)
rlLevelOne.explorationProb = 0
levelOnePolicy = rlLevelOne.getAction
def oppLevelOnePolicy(state):
    return [levelOnePolicy((state[1], state[0]))[0]]
def ownLevelOnePolicy(state):
    return [levelOnePolicy(state)[0]]

rlLevelTwo = QLearning(rlalg, oppLevelOnePolicy, identityFeatureExtractor, maxIter = maxIter)
rlLevelTwo.explorationProb = 0
levelTwoPolicy = rlLevelTwo.getAction
def oppLevelTwoPolicy(state):
    return [levelTwoPolicy((state[1], state[0]))[0]]
def ownLevelTwoPolicy(state):
    return [levelTwoPolicy(state)[0]]
```
